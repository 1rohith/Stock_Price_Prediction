```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('C:/Users/vrohi/Downloads/Apple/aapl_2014_2023.csv')
```

```python
df['date']=pd.to_datetime(df['date'])
df.sort_values('date', inplace=True)
```

```python
x=np.array(df.index).reshape(-1,1)
y=df['open'].values
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df.index, df['open'], color='green', label='Opening Stock')
plt.plot(df.index, model.predict(x), color='red', linestyle='dashdot', label='Closing Stock')
plt.title("APPLE Inc. Stock Price(2014-2023)")
plt.xlabel('open')
plt.ylabel('close')
plt.legend()
plt.show()
```
